```python
from ultralyticsplus import YOLO, render_result

# load best model weights 'best.pt' @ Hugging Face
model = YOLO('flkennedy/YOLOv8s-building-segmentation-demo')

# set image path here, jpeg format definitely works.
# upload using the 'Files' area on the left bar of the notebook
image = 'YOUR-IMAGE-FILENAME.jpeg'

# prediction on your image
results = model.predict(image)

render = render_result(model=model, image=image, result=results[0])
render
```
